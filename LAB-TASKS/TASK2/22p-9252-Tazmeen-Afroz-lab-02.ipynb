PROBLEM-1 : Question: Analyzing a Social Network Dataset
You are given a dataset representing a social network where nodes represent individuals, and edges
represent connections between individuals. Additionally, everyone has attributes such as age and number
of connections. The dataset is provided in CSV format.
1. Data Loading:
• Load the dataset into Pandas Data Frame.
• Check for missing values and handle them appropriately.
2. Social Network Analysis:
• Use NetworkX to create a graph representation of the social network.
3. Histogram Creation:
• After creating the social network graph, visualize the relationship between age and
connection strength by generating a histogram. Plot age on one axis and connection
strength on the other. This will provide insights into the distribution of connections across
different age groups in the social network.

```python
import numpy as np 
import pandas as pd 
import networkx as nx 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
```

```python
missing_values = df.isnull().sum()
```

```python
missing_values
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df['connections'].fillna(df['connections'].mean() ,inplace=True)
```

```python
df.isnull().sum()
```

```python
G = nx.from_pandas_edgelist(df)
```

```python
G.nodes
```

```python
G.edges
```

```python
nx.draw(G)
```

```python
options = {
    
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "blue",
    "linewidths":4,
    "with_labels": True
}
```

```python
nx.draw(G, **options)
```

```python
plt.hist(df['age'],label='age',alpha=0.9,edgecolor='black')

plt.hist(df['connections'], label='connections', alpha= 0.9, edgecolor='black')
plt.legend()
plt.title('Showing the frequency of ages and connections')
plt.xlabel('Age and connections')
plt.ylabel('Frequency')
plt.show()
```

1. Exploring the Standard Normal Distribution in Python:
a. Generate a set of at least 1000 random numbers that follow a standard normal distribution
(with a mean of 0 and a standard deviation of 1).
b. Create a histogram to visually represent the distribution of the generated random numbers.
c. Calculate and display both the sample mean and sample standard deviation of your
generated dataset. Compare these values to the expected mean and standard deviation for
a standard normal distribution.

```python
normal_values = np.random.normal(loc=0.0,scale = 1.0,size = 1000)
```

```python
plt.hist(normal_values,bins= 20,edgecolor='black',color='grey')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Normal Values')
plt.show()
```

```python
mean = np.mean(normal_values)
std = np.std(normal_values)
print(f'Sample mean is: {mean}')
print(f'Sample std is:{std}')
```

```python
#Comparison
#Expected mean is 0 , Expected std is 1 
expected_mean = 0
expected_std = 1

print(f'Difference btw my sample mean and expected mean is:{mean - expected_mean}')
print(f'Difference btw my sample std and expected std is :{std-expected_std}')
```

```python
print('The sample mean and sample std are near the expected mean and std.')
```

Write a Python program to plot two or more lines and set the line markers. The code snippet gives the
output shown in the following screenshot:

```python
x1 = np.linspace(0,10,10)
y1 = 2*x1 + 1

x2 = np.linspace(0,10,10)
y2 = np.cos(x2)

fig ,ax = plt.subplots()

ax.plot(x1,y1,'-o',label=" y = 2x+1")
ax.plot(x2,y2,'-x',label=" y = cosx")

ax.legend()
ax.set_title("Straight Line and Cosine")
ax.set_xlabel('x-axis')
ax.set_ylabel('y_axis')

plt.show()
```
