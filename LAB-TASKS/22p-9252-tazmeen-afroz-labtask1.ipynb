```python
class Student:
    def __init__(self,roll_number,name,marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

    
```

```python
student_database = []
```

```python
def add_student(roll_number, name, marks):
 student_database.append(Student(roll_number,name,marks))
```

```python
def display_student_details(roll_number):
    check= False
    for student in student_database:
        if student.roll_number == roll_number:
         check = True
         print("Student Name is",student.name)
         print(" Student Roll_No is ",student.roll_number)
         print("Student marks are ",student.marks)
        
    if(check == False):
     print("Student Not found")
```

```python
def display_all_students():
    for student in student_database:
        print("Student Name is",student.name)
        print(" Student Roll_No is ",student.roll_number)
        print("Student marks are ",student.marks)
    print("\nNo more data")
```

```python
def calculate_average_marks(roll_number):
    check= False
    avg=0
    for student in student_database:
        check = True
        if student.roll_number == roll_number:
              avg = sum(student.marks.values())/len(student.marks.values())
    if(check == False):
     print("Student Not found")
    return avg

```

```python
def find_topper():
    top = 0
    toppers = []
    
    for student in student_database:
        avg = sum(student.marks.values()) / len(student.marks.values())
        if avg > top:
            top = avg
            toppers = [student.name]
        elif avg == top:
            toppers.append(student.name)

    return toppers

        
```
